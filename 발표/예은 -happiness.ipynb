**데이터의 column name정리**

```python
data_2015['Year']=2015
data_2016['Year']=2016
data_2017['Year']=2017
data_2018['Year']=2018
data_2019['Year']=2019
```

```python
# get column name : 'Country', 'Region', 'Happiness Rank', 'Happiness Score',
#     'Standard Error', 'Economy (GDP per Capita)', 'Family',
#        'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
#        'Generosity', 'Dystopia Residual', 'Year'
data_2015.columns
data_2015=data_2015.rename(columns = {'Happiness Rank':'Rank','Happiness Score':'Score','Economy (GDP per Capita)':'GDP per Capita',
                           'Health (Life Expectancy)':'Life Expectancy','Trust (Government Corruption)':'Government Corruption'})
data_2015
```

```python
# 'Country', 'Region', 'Happiness Rank', 'Happiness Score',
#        'Lower Confidence Interval', 'Upper Confidence Interval',
#        'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
#        'Freedom', 'Trust (Government Corruption)', 'Generosity',
#        'Dystopia Residual', 'Year'
data_2016.columns
data_2016=data_2016.rename(columns={'Happiness Rank':'Rank','Happiness Score':'Score','Economy (GDP per Capita)':'GDP per Capita',
                          'Health (Life Expectancy)':'Life Expectancy','Trust (Government Corruption)':'Government Corruption'})
data_2016
```

```python
# 'Country', 'Happiness.Rank', 'Happiness.Score', 'Whisker.high',
#        'Whisker.low', 'Economy..GDP.per.Capita.', 'Family',
#        'Health..Life.Expectancy.', 'Freedom', 'Generosity',
#        'Trust..Government.Corruption.', 'Dystopia.Residual', 'Year'
data_2017.columns
data_2017=data_2017.rename(columns={'Happiness.Rank':'Rank','Happiness.Score':'Score','Economy..GDP.per.Capita.':'GDP per Capita',
                         'Health..Life.Expectancy.':'Life Expectancy','Trust..Government.Corruption.':'Government Corruption',
                         'Dystopia.Residual':'Dystopia Residual'})
data_2017
```

```python
# 'Overall rank', 'Country or region', 'Score', 'GDP per capita',
#        'Social support', 'Healthy life expectancy',
#        'Freedom to make life choices', 'Generosity',
#        'Perceptions of corruption', 'Year'
data_2018.columns
data_2018 = data_2018.rename(columns={'Overall rank':'Rank','Country or region':'Country','Healthy life expectancy':'Life expectancy',
                          'Perceptions of corruption':'Government Corruption','Freedom to make life choices':'Freedom'})
data_2018
```

```python
# 'Overall rank', 'Country or region', 'Score', 'GDP per capita',
#        'Social support', 'Healthy life expectancy',
#        'Freedom to make life choices', 'Generosity',
#        'Perceptions of corruption', 'Year'
data_2019.columns
data_2019=data_2019.rename(columns={'Overall rank':'Rank','Country or region':'Country','Healthy life expectancy':'Life expectancy',
                         'Freedom to make life choices':'Freedom','Perceptions of corruption':'Government Corruption'})
data_2019
```

2015,2016 데이터 기반으로 2017,2018,2019 Country의 Region추가

```python
region_list = pd.DataFrame(data_2015.loc[:,['Country','Region']])
region_list
```

```python
region_list2 = pd.DataFrame(data_2016.loc[:,['Country','Region']])
region_list2
```

```python
region_data=pd.merge(region_list,region_list2,how='outer')
region_data
```

```python
data_2017=data_2017.merge(region_data,how='left')
data_2017
```

```python
data_2018=data_2018.merge(region_data,how='left')
data_2018
```

```python
data_2019=data_2019.merge(region_data,how='left')
data_2019
```
