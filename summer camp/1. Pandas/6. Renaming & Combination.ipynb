# 6. Renaming

```python
import pandas as pd
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris
```

### Renaming
- rename() function으로 할 수 있다.


```python
# column변경
iris.rename(columns = {'sepal_length' : 's_l'})
```

```python
# index 변경
iris.rename(index = {0 : 'zero'})
```

```python
# axis rename
iris.rename_axis("nums", axis='rows').rename_axis("fields", axis='columns')
```

### Combining
- concat() : 위, 아래로 간단하게 붙이기(row가 늘어난다)
- join() : 옆으로 붙이기(index를 중심으로 joing)



```python
# Concat 예시
double_iris = pd.concat([iris, iris])
double_iris
```
