# Grouping and sorting

```python
import pandas as pd
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris
```

### Gropuwize anylisys


```python
#iris.groupby('species')
#iris.groupby('species').species.count()
#iris.groupby('species').sepal_length.min()
iris.groupby('species').apply(lambda df: df.loc[0].sepal_length + df.loc[0].petal_width)
```

```python
# 실행되지 않는 예시지만 이런것도 가능
#reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
iris.groupby(['species']).sepal_length.agg([len, min, max])
```

### Multi indexes
- groupby() 연산을 하면 multi index 도 나올 수 있다.

```python
multi = iris.groupby(['species', 'petal_width']).sepal_length.agg([len])
multi
```

```python
# reset_index()로 정리할 수 있다
multi = multi.reset_index()
multi
```

### Sorting
- 값으로 sorting ( + 역순)
- index로 sorting
- 복수의 기준으로

```python
#multi.sort_values(by='len')
mixed_index = multi.sort_values(by='len', ascending = False)
```

```python
# sorting by index
mixed_index.sort_index()
```

```python
# sort by more than one column
multi.sort_values(by=['petal_width', 'len'])
```
