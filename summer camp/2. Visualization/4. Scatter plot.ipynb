```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Data load
- 보험료 데이터

```python
path = "insurance.csv"
insurance_data = pd.read_csv(path)
insurance_data.head()
```

### Scatter plot 그리기
 - scatterplot()

```python
sns.scatterplot(data = insurance_data , x = 'bmi', y = 'charges')
```

### regplot
- regression line 추가해서 그리기(두 변수간의 관계를 설명하는 y = ax + b 식)

```python
sns.regplot(data = insurance_data , x = 'bmi', y = 'charges')
```

### Group을 지어서 scatterplot 그리기

```python
sns.scatterplot(data = insurance_data , x = 'bmi', y = 'charges', hue='smoker')
```

### Line여러개 추가

```python
sns.lmplot(x = 'bmi', y = 'charges', hue = 'smoker', data = insurance_data)
```

### class 로 분활된 그래프 그리기
- swarmplot()

```python
sns.swarmplot(x = 'smoker', y = 'charges', data = insurance_data)
```
