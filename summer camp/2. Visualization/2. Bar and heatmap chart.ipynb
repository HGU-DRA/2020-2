```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Barchart
- 항목별로 분류된것의 value를 표현할 때 자주쓰인다
- sns.barplot()

#### Data load

```python
sns.set(style = "whitegrid")
tips = sns.load_dataset("tips")
tips.head()
```

- day에 따른 total bill dml 평균구하기

```python
plt.figure(figsize= (10,6))
plt.title("Averate total_Bill per day")
sns.barplot(x  = tips.day, y=tips.total_bill)
```

- Group by 연산하기

```python
plt.figure(figsize= (10,6))
plt.title("Averate total_Bill per day")
sns.barplot(x = "day", y="total_bill", hue = "sex", data = tips)
```

### Heat map
- 변수들간의 상관관계를 나타낼때 주로 쓰인다

#### Data load

```python
import numpy as np; np.random.seed(0)
uniform_data = pd.DataFrame(np.random.rand(10, 12))
uniform_data
```

```python
ax = sns.heatmap(uniform_data)
```

#### Data load(항공데이터)

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

```python
sns.heatmap(flights)
```

 - 적용할 수 있는 다양한 옵션들
    - fmt, linewidth,cmap, center

```python
sns.heatmap(flights, annot = True, fmt = "d" )
```

```python
ax = sns.heatmap(flights, linewidths=.5)
```

```python
ax = sns.heatmap(flights, cmap="YlGnBu")
```

```python
ax = sns.heatmap(flights, center=flights.loc["January", 1955])
```
