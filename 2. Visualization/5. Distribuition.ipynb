```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Load iris data

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris
```

### Draw histogram (distribution plot)
- distplot(a = "")

```python
sns.distplot(a= iris['sepal_length'], kde = False)
#kde 를 true로 하면 밀도그래프가 나온다 나온다.(kernal density estimation)
```

### kde plot(kernel distribution plot)

```python
sns.kdeplot(data=iris['petal_length'],shade=True, hue = 'species')
```

### 2D kde plot

```python
#2D kde_plot
sns.jointplot(x = iris['petal_length'], y = iris['sepal_length'], kind = 'kde')
```

### Group by, dist plot and kde plot

```python
iris.species.unique()
```

```python
iris_setosa = iris.loc[iris.species == 'setosa']
iris_versicolor = iris.loc[iris.species == 'versicolor']
iris_virginica = iris.loc[iris.species == 'virginica']
```

```python
sns.distplot(a = iris_setosa['petal_length'],label = "setosa", kde = False)
sns.distplot(a = iris_versicolor['petal_length'],label = "versicolor", kde = False)
sns.distplot(a = iris_virginica['petal_length'],label = "virginica", kde = False)

plt.title("Histogram of petal lengths, by species")
plt.legend()
```

```python
sns.kdeplot(data = iris_setosa['petal_length'],label = "setosa", shade = True)
sns.kdeplot(data = iris_versicolor['petal_length'],label = "versicolor", shade = True)
sns.kdeplot(data = iris_virginica['petal_length'],label = "virginica", shade = True)

plt.title("Histogram of petal lengths, by species")
plt.legend()
```
