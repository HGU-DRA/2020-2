```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Data load
fmri데이터를 로드한다.

```python
fmri = sns.load_dataset("fmri")
fmri['signal2'] = (fmri.signal+1)/2
```

```python
#fmri.head()
fmri.region.unique()
```

### draw plane linepot
-  sns.lineplot()

```python
sns.lineplot(data = fmri, x = "timepoint", y ="signal")
```

- option추가하기

```python
plt.figure(figsize = (14,6))
plt.title("fmri signal data")
sns.lineplot(data = fmri, x="timepoint",  y = "signal")
```

- 여러개 한번에 그리기
- label추가(범례)

```python
plt.figure(figsize = (14,6))
plt.title("fmri signal data")
sns.lineplot(data = fmri, x="timepoint",y="signal", label = "Signal")
sns.lineplot(data = fmri, x="timepoint",y="signal2", label = "Signal2")
plt.xlabel('Time point')
```

- Group by 사용하기

따로 연산을 할 필요 없이, hue = "", style = "" 안에 group지을 column명을 넣어주면 된다

```python
fmri.event.unique()
# fmir의 event에는 stim 과 cue 항목이 있다.
```

```python
sns.lineplot(data = fmri, x = "timepoint", y="signal", hue = "event")
```

- 특정부분만 선택한 그래프를 그릴수도있다. 아래 예제는 region이 frontal인 부분만 가져온것

```python
sns.lineplot(data = fmri.query("region == 'frontal'"), x ="timepoint",
            y = "signal", hue = "event")
```

- 위 그래프들은 평균값과 최고, 최저 를 나타낸 그래프이다. 만약 전체 line을 그리고 싶으면 그릴 수 있다.

```python
sns.lineplot(x="timepoint", y="signal", hue="event",
                  units="subject", estimator=None, lw = 1,
                  data=fmri.query("region == 'frontal'"))
```
