## Indexing in pandas
- loc 과 iloc
- iloc : integer based selection(position based)
- loc : label-based selection(not postion based)

```python
iris.iloc[0]
```

```python
iris.iloc[:,0]
```

```python
iris.iloc[:3,0]
```

```python
iris.iloc[[0,1,2],0]
```

```python
iris.iloc[-5:]
```

```python
#iris.iloc[0,0]#error
iris.loc[0,'sepal_length']
```

#### 중요한 차이점
iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.



```python
iris.iloc[0:10] #0<= <10
#iris.loc[0:10] #이름? <  >
```

### Conditional selection

```python
iris.species == 'setosa'
```

```python
iris.loc[iris.species == 'setosa'] #iloc은 안됨
```

```python
iris.loc[(iris.petal_width > 0.3) & (iris.species == 'setosa')] #iloc은 안됨
```

```python
iris.loc[(iris.petal_width > 0.3) | (iris.species == 'setosa')] #iloc은 안됨
```

```python
iris.loc[iris.species.isin(['setosa','virginica'])] #iloc은 안됨
```

```python
iris.loc[iris.species.notnull()] #iloc은 안됨
```
